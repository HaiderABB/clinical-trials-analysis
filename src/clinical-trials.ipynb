checking if the location of the organization and the location of the facility are the same to see correlation 

```python
from rapidfuzz import fuzz
import re
import pandas as pd

# Cleaning function
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # normalize spaces
    return text


def fuzzy_match_and_replace(row, threshold=85):
    org = row['organization_fullname']
    loc = row['location_facility']
    
    if pd.isna(org) or pd.isna(loc):
        return row  # skip if any is NaN

    org_clean = clean_text(org)
    loc_clean = clean_text(loc)
    
    score = fuzz.token_set_ratio(org_clean, loc_clean)
    
    if score >= threshold:
        # Choose the longer original string (not cleaned)
        unified_name = org if len(org) >= len(loc) else loc
        row['organization_fullname'] = unified_name
        row['location_facility'] = unified_name
        row['is_same'] = True
    else:
        row['is_same'] = False

    return row

# Apply to the whole DataFrame
df = df.apply(fuzzy_match_and_replace, axis=1)

# View subset
subset = df[['organization_fullname', 'location_facility', 'is_same']]
subset.head(30)
```

```python
import re

# Your date columns
date_cols = [
    'completion_date',
    'last_update_post_date',
    'primary_completion_date',
    'start_date',
    'study_first_post_date',
    'study_first_submit_date'
]

# Function to insert day=15 if day is missing
def fill_day_15(date_str):
    if pd.isna(date_str):
        return np.nan
    date_str = str(date_str).strip()
    
    # Check if format is like "Jan 2023" or "2023-01"
    if re.fullmatch(r'[A-Za-z]{3,9}\s+\d{4}', date_str):  # e.g., "March 2021"
        return f'15 {date_str}'
    elif re.fullmatch(r'\d{4}-\d{2}', date_str):  # e.g., "2023-01"
        return f'{date_str}-15'
    return date_str  # return unchanged if full date is already present

# Apply transformation and convert to datetime
for col in date_cols:
    df[col] = df[col].apply(fill_day_15)  # Fill in day if needed
    df[col] = pd.to_datetime(df[col], errors='coerce')  # Convert to datetime
```

Checking for correlation between people enrolled and the status of the trial

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Convert to numeric and handle missing
df['enrollment'] = pd.to_numeric(df['enrollment'], errors='coerce')
df['enrollment'].fillna(0, inplace=True)

# Filter for only relevant statuses
filtered_df = df[df['overall_status'].isin(['COMPLETED', 'TERMINATED'])]

# Remove outliers using IQR method for each group
def remove_outliers_iqr(group):
    Q1 = group['enrollment'].quantile(0.25)
    Q3 = group['enrollment'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group['enrollment'] >= lower_bound) & (group['enrollment'] <= upper_bound)]

# Apply outlier removal
filtered_df = filtered_df.groupby('overall_status', group_keys=False).apply(remove_outliers_iqr)

# Boxplot after removing outliers
plt.figure(figsize=(8, 5))
sns.boxplot(data=filtered_df, x='overall_status', y='enrollment', palette='Set2')
plt.title('Enrollment Distribution by Trial Status (Outliers Removed)')
plt.ylabel('Number of People Enrolled')
plt.xlabel('Trial Status')
plt.tight_layout()
plt.show()

# Summary statistics
summary = filtered_df.groupby('overall_status')['enrollment'].describe()
print("\nSummary statistics:")
print(summary)

# T-test: Compare mean enrollments
completed = filtered_df[filtered_df['overall_status'] == 'COMPLETED']['enrollment']
terminated = filtered_df[filtered_df['overall_status'] == 'TERMINATED']['enrollment']

t_stat, p_value = ttest_ind(completed, terminated, equal_var=False)
print(f"\nT-test Results: t-statistic = {t_stat:.3f}, p-value = {p_value:.5f}")

# Verdict
if p_value < 0.05:
    print("\n✅ There is a **statistically significant** difference in enrollment between COMPLETED and TERMINATED trials.")
else:
    print("\n❌ There is **no statistically significant** difference in enrollment between COMPLETED and TERMINATED trials.")
```

Checking for trend in lead sponsors and its relation with 'COMPLETED'/'TERMINATED'/'RECRUITING'

```python
# Count number of trials per sponsor
sponsor_counts = df['lead_sponsor'].value_counts()

# Keep sponsors with more than 100 trials
top_sponsors = sponsor_counts[sponsor_counts > 50].index

# Filter DataFrame for top sponsors only
top_df = df[df['lead_sponsor'].isin(top_sponsors)]

# Group by sponsor and trial status
status_counts = top_df.groupby(['lead_sponsor', 'overall_status']).size().unstack(fill_value=0)

# Compute total trials
total_trials = status_counts.sum(axis=1)

# Compute percentage for COMPLETED and TERMINATED only
percent_df = status_counts[['COMPLETED', 'TERMINATED']].div(total_trials, axis=0) * 100

# Rename columns for clarity
percent_df = percent_df.rename(columns={
    'COMPLETED': 'COMPLETED_%',
    'TERMINATED': 'TERMINATED_%'
})

# Sort by total number of trials (optional)
percent_df = percent_df.loc[percent_df.index, ['COMPLETED_%', 'TERMINATED_%']]

# Sort by COMPLETED_% in descending order
percent_df_sorted = percent_df.sort_values(by='COMPLETED_%', ascending=False)

print(percent_df_sorted)

```
